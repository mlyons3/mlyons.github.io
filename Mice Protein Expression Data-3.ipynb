### <font color=purple >  Exploratory Analysis </font>


As stated above, we will begin by taking the mean of the 15 measurements of protein expression for each mouse. First, however, a bit more manipulation of the data is necessary. The MouseID column contains two bits of information -- the mouse that was experimented on, and the number of the measurement that any particular row represents (i.e. 309_1 denotes the first measurement taken on mouse 309). We split this column into two columns -- one for the mouse and one for the measurement number -- so that we could then easily group by mouse and take the average of the protein expression data over the 15 trials. 


```python
# Sorting each measurement of protein expression in each mouse expression
df_proteins["Mouse"] = df_proteins["MouseID"].str.split("_", expand=True)[0]
df_proteins["Measurement#"] = df_proteins["MouseID"].str.split("_", expand=True)[1]
df_proteins.drop(columns="MouseID", inplace=True)
df_proteins = df_proteins.set_index(["Genotype", "Treatment", "Behavior", "Measurement#"])

#Displaying Dataframe 
df_proteins
```

```python
# Obtaining the mean protein expression for each mouse in each class 
mean_expressions = df_proteins.groupby("Mouse").mean()
mean_expressions
```

####  <font color=green > Explaining the Data </font>

The data that we will be using in this project only pertains to the control mice, which are normal mice with no underlying health issues. We will not be using the Ts65Dn (down syndrome) mice in our analysis. There are several other variables including treatment and learning. A mouse can either be treated by memantine, which is a drug used to treat alzheimer's disease, or saline, which has no effect and is used as the control. Mice can also be stimulated to learn from context-shock associative learning, or not stimulated to learn from shock-context associative learning. So we will be working with 4 classes of mice: 
- Control-Context Shock-Saline (c_CS_s) 
- Control-Context Shock-Memantine (c_CS_m)
- Control-Shock Context-Saline (c_SC_s)
- Control-Shock Context-Memantine (c_SC_m) 

In each of these 4 classes of mice, there are 77 protein whose expression level in the nuclear fraction of the cortex, is measured. The amoutn of expression is already standardized across all classes and mice.  

We will begin by examining the differences in protein expression for the control mice who were stimulated to learn when given either saline or memantine. The two classes of interest are thus c-CS-s and c-CS-m.

```python

control_stim_saline = df_proteins[df_proteins["class"] == "c-CS-s"]
control_stim_mem = df_proteins[df_proteins["class"] == "c-CS-m"]
control_stim_saline
```

Now, we just want the mice in the c-CS-s and c-CS-m classes, since this will allow us to index into the mean_expressions DataFrame and retrieve only the relevant data. 

```python
saline_mice = list(control_stim_saline["Mouse"].unique())
memantine_mice = list(control_stim_mem["Mouse"].unique())

#Grab the rows from the mean_expressions DataFrame corresponding to the mice in these two classes.
saline_expressions = mean_expressions.loc[saline_mice][df_proteins.columns[:20]]
memantine_expressions = mean_expressions.loc[memantine_mice][df_proteins.columns[:20]]

# Displaying example data frame of the average protein expression for each protein in each mouse 
memantine_expressions
```

Now we're going to take a step back in order to analyze some of the proteins. As seen in the heatmaps above, protein expression for the each protein varies slightly across each mice. To get a better feel for the protein expression per class, we are going to take an average of protein expression across all mice per class for each protein. We will then display the results in a series of heatmaps in order to find which proteins stand out. We will split the proteins between two heatmaps since there are a large number of proteins. 

```python
# Creating dataframes for control non-stimulated learning because stimulated learning is already made 

control_stim_saline = df_proteins[df_proteins["class"] == "c-CS-s"]
saline_mice_stim = list(control_stim_saline["Mouse"].unique())

control_stim_memantine = df_proteins[df_proteins["class"] == "c-CS-m"]
memantine_mice_stim = list(control_stim_memantine["Mouse"].unique())

control_nostim_saline = df_proteins[df_proteins["class"] == "c-SC-s"]
saline_mice_nostim = list(control_nostim_saline["Mouse"].unique())


control_nostim_mem = df_proteins[df_proteins["class"] == "c-SC-m"]
memantine_mice_nostim = list(control_nostim_mem["Mouse"].unique())



# c-CS-s
c_CS_s = mean_expressions.loc[saline_mice_stim][df_proteins.columns[:len(df_proteins.columns) - 2]]

# c-CS-m
c_CS_m = mean_expressions.loc[memantine_mice_stim][df_proteins.columns[:len(df_proteins.columns) - 2]]

# c-SC-s
c_SC_s = mean_expressions.loc[saline_mice_nostim][df_proteins.columns[:len(df_proteins.columns) - 2]]

# c-SC-m
c_SC_m = mean_expressions.loc[memantine_mice_nostim][df_proteins.columns[:len(df_proteins.columns) - 2]]

# Finding mean of each protein in each class
c_CS_s_MEAN = c_CS_s.mean()
c_CS_m_MEAN = c_CS_m.mean()
c_SC_s_MEAN = c_SC_s.mean()
c_SC_m_MEAN = c_SC_m.mean()


# Making into data frame
frame = {"c_CS_s": c_CS_s_MEAN, "c_CS_m": c_CS_m_MEAN, "c_SC_s": c_SC_s_MEAN, "c_SC_m": c_SC_m_MEAN}  
meanControl = pd.DataFrame(frame)
meanControl1 = meanControl[:38].T
meanControl2 = meanControl[38:].T

# Plotting first 38 proteins
plt.subplots(figsize=(15,10))
plt.title("Heatmap for First 38 Proteins based on Class and Protein Expression", fontsize =20)
sns.heatmap(meanControl1, cmap="YlGnBu")
```

```python
# Plotting last 39 proteins 
plt.subplots(figsize=(15,10))
plt.title("Heatmap for Last 39 Proteins based on Class and Protein Expression", fontsize =20)
sns.heatmap(meanControl2, cmap="YlGnBu")
plt.xlabel("Protein", fontsize = 20)
plt.ylabel("Class", fontsize = 20)
```
